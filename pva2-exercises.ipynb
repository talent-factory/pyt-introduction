Wir haben einige Funktionen wie `sum()`, `len()` oder `str()` kennengelernt. In diesem Zusammenhang ist es wichtig , deren Funktionalität jeweils auch in Handbuch nachzulesen, um ihre vollen Möglichkeiten auszuschöpfen.

In der folgenden Aufgabe wollen wir eine Zeichenkette (`<class 'str'>`) analysieren. Somit benötigen wir Wissen, welche Funktionen uns diese Klasse zur Verfügung stellt. Die entsprechende Methoden sind [hier](https://docs.python.org/3.12/library/stdtypes.html?highlight=split#string-methods) beschrieben.

**Aufgabe:** Erstellen Sie eine Liste von Wörtern aus einem gegebenen Satz und zähle, wie oft jedes Wort vorkommt (_Hinweis:_ [split()](https://docs.python.org/3.12/library/stdtypes.html?highlight=split#str.split)).

```python
# Definiert den gegebenen Satz
satz = "apple banana apple strawberry banana lemon"

# Erstellt eine Liste von Wörtern durch Trennen des Satzes
wortliste = satz.split()

# Erzeugt ein leeres Wörterbuch (dictionary)
wortzaehler = {}

# Gehe jedes Wort in der Liste durch
for wort in wortliste:
    # Wenn dies das erste Mal ist, dass wir dieses Wort sehen,
    # füge es zum Wörterbuch hinzu mit dem Wert 1
    if wort not in wortzaehler:
        wortzaehler[wort] = 1
    # Wenn wir das Wort bereits gesehen haben, erhöhe seinen Zähler um 1
    else:
        wortzaehler[wort] += 1

# Drucken das Wörterbuch zum Überprüfen
print(wortzaehler)
```

Dieses Skript zählt die Wörter in dem gegebenen Satz und speichert die Werte in einem Dictionary für einfachen Zugriff. Jedes Mal wenn es ein Wort trifft, das es schon mal gesehen hat, erhöht es den Zähler um eins. Wenn es ein neues Wort trifft, fügt es dieses Wort zum Wörterbuch hinzu und setzt seinen Zähler auf eins. Es benutzt die split() Methode um den Satz in eine Liste von Wörtern zu zerlegen. Die split() Methode trennt einen String an Leerzeichen, wenn kein Trennparameter angegeben ist.

## Aufgabe 2
Sie haben ein Wörterbuch mit Stundensätzen für verschiedene Aufgaben.

```python
stundensatz = {
    "washing": 10,   # Stundensatz (CHF) für das Waschen
    "cleaning": 20,  # ...   
    "cooking": 15
}
```

**Aufgabe:** Basierend auf diesen Stundensätzen wurden die folgenden Aufgaben erldigt. Berechne die Gesamtkosten für verschiedene Mengen von Aufgaben.

```python
aufgaben = {
    "washing": 2,  # Für diese Aufgabe wurden 2 Stunden benötigt
    "cleaning": 3, # ...
    "cooking": 4
}

# Initialisiere die Gesamtkosten mit 0
total_kosten = 0

# Gehe durch jede Aufgabe in aufgaben
for aufgabe, stunden in aufgaben.items():
    # Multipliziere die Anzahl der Stunden für diese Aufgabe mit dem entsprechenden Stundensatz und addiere die Kosten zur Gesamtsumme
    total_kosten += stunden * stundensatz[aufgabe]

print(f'Total Kosten: CHF {total_kosten}.--')
```
